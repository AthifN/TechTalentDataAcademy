```python
#Find the mean number of all inclusive hotels across all destinations.
DestData['All_Inclusive_Hotel_Count'].mean()
```

```python
#Find the lowest scoring destination.
DestData[DestData.Avg_Destination_Rating == DestData.Avg_Destination_Rating.min()]
```

```python
#Find the highest scoring destination.
DestData[DestData.Avg_Destination_Rating == DestData.Avg_Destination_Rating.max()]
```

```python
#Find all the destinations where there are more than 9 all inclusive hotels.

DestData[DestData.All_Inclusive_Hotel_Count >= 9]
```

```python
#Filter the data by score above 8.

DestData[DestData.Avg_Destination_Rating >= 8]
```

```python
#Filter the data score below 2 ( I need to know if these destinations should be removed or there is a problem)

DestData[DestData.Avg_Destination_Rating <= 2]
```
